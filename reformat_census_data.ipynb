```python
import geopandas as gpd
import pandas as pd
import shapely
```

```python
census = pd.DataFrame(gpd.read_file('tabblock2010_51_pophu.shp'))
```

```python
census_url = 'https://github.com/ZackBoner/CS-4774-Project/raw/main/nhgis0002_ds172_2010_block.zip'
census_block_data = pd.read_csv(census_url)
```

```python
census_columns = ['STATE', 'COUNTY', 'COUNTYA', 'PLACEA', 'TRACTA', 'BLKGRPA', 'BLOCKA', 'H7X001', 'H7X002', 'H7X003', 'H7X004', 'H7X005', 'H7X006', 'H7X007', 'H7X008']
census_block_data = census_block_data[census_columns]
```

```python
census_block_data.sort_values('BLOCKA')
```

```python
census.sort_values('BLOCKCE')
```

```python
census.BLOCKCE = census.BLOCKCE.astype('int')
census_block_data.BLOCKA = census_block_data.BLOCKA.astype('int')
census.COUNTYFP10 = census.COUNTYFP10.astype('int')
census_block_data.COUNTYA = census_block_data.COUNTYA.astype('int')
```

```python
census = census.sort_values(['COUNTYFP10', 'TRACTCE10', 'BLOCKCE'])
census_block_data = census_block_data.sort_values(['COUNTYA', 'TRACTA', 'BLOCKA'])
```

```python
census = census.rename(columns={'BLOCKCE': 'BLOCK', 'COUNTYFP10': 'COUNTYA', 'TRACTCE10': 'TRACTA'})
census_block_data = census_block_data.rename(columns={'BLOCKA': 'BLOCK'})
```

```python
census_block_data = census_block_data.set_index(['COUNTYA', 'TRACTA', 'BLOCK'])
```

```python
census = census.set_index(['COUNTYA', 'TRACTA', 'BLOCK'])
```

```python
census_data_final = pd.concat([census.reset_index(drop=True), census_block_data.reset_index(drop=True)], axis=1)
```

```python
census_data_final.drop(['POP10', 'PLACEA', 'PARTFLG', 'HOUSING10', 'STATEFP10'], axis=1, inplace=True)
```
