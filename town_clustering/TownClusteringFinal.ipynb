```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import pickle
%matplotlib inline
```

```python
towns = pd.read_pickle("final_combined_dataset.pkl")
```

```python
towns.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42).fit(towns[["closest_solid_waste", "closest_renewable_energy", "closest_petroleum_release", "closest_pollution_reports", "median income", "bachelors degree", "white", "black", "hispanic", "asian"]])
towns["cluster"] = kmeans.labels_
```

```python
import plotly.express as px
# import geopandas as gpd

# geo_df = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))

# px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_geo(towns,
                    lat=towns["latitude"],
                    lon=towns["longitude"],
                    color = towns['cluster'],
                    hover_name="Town")
fig.update_layout(
    geo = dict(
        scope = 'north america',
        showland = True,
        landcolor = "rgb(212, 212, 212)",
        subunitcolor = "rgb(255, 255, 255)",
        countrycolor = "rgb(255, 255, 255)",
        showlakes = True,
        lakecolor = "rgb(255, 255, 255)",
        showsubunits = True,
        showcountries = True,
        resolution = 50,
        projection = dict(
            type = 'conic conformal',
            rotation_lon = -100
        ),
        lonaxis = dict(
            showgrid = True,
            gridwidth = 0.5,
            range= [ -140.0, -55.0 ],
            dtick = 5
        ),
        lataxis = dict (
            showgrid = True,
            gridwidth = 0.5,
            range= [ 20.0, 60.0 ],
            dtick = 5
        )
    ),
)

fig.show()
```
