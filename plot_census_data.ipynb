```python
import plotly
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
cdf = pd.read_pickle('processed_pickles/census_data.pkl.xz')
```

```python
gcdf = gpd.GeoDataFrame(cdf)
```

```python
gcdf_pos_pop = gcdf[gcdf.H7X001 > 0]
```

```python
gcdf_pos_pop
```

```python
gcdf_pos_pop['prop_white'] = gcdf_pos_pop['H7X002']/gcdf_pos_pop['H7X001']
gcdf_pos_pop['prop_black'] = gcdf_pos_pop['H7X003']/gcdf_pos_pop['H7X001']
gcdf_pos_pop['prop_asian'] = gcdf_pos_pop['H7X004']/gcdf_pos_pop['H7X001']
gcdf_pos_pop['prop_hispanic'] = gcdf_pos_pop['H7X007']/gcdf_pos_pop['H7X001']
```

```python
gcdf_pos_pop
```

```python
gcdf_pos_pop.plot(column='prop_white', figsize=(20, 20), cmap='magma', legend=True, legend_kwds={'label': 'proportion white', 'orientation': 'horizontal'});
```

```python
gcdf_pos_pop.plot(column='prop_black', figsize=(20, 20), cmap='magma', legend=True, legend_kwds={'label': 'proportion black', 'orientation': 'horizontal'});
```

```python
gcdf_pos_pop.plot(column='prop_asian', figsize=(20, 20), cmap='magma', legend=True, legend_kwds={'label': 'proportion asian', 'orientation': 'horizontal'});
```

```python
gcdf_pos_pop.plot(column='prop_hispanic', figsize=(20, 20), cmap='magma', legend=True, legend_kwds={'label': 'proportion hispanic', 'orientation': 'horizontal'});
```
